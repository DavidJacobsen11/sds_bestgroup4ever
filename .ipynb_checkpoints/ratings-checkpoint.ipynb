# Subfunctions

```python
def find_budget(soup):
    for i in soup.find(id="titleDetails").find_all("div", class_="txt-block"):
        if i.find().text == "Budget:":
            return float(i.text.strip().split("\n")[0].split("$")[1].replace(",", ""))
        
def find_gross(soup):
    for i in soup.find(id="titleDetails").find_all("div", class_="txt-block"):
        if i.find().text == "Cumulative Worldwide Gross:":
            return float(i.text.strip().split("\n")[0].split("$")[1].replace(",", ""))
        
def find_grossUSA(soup):
    for i in soup.find(id="titleDetails").find_all("div", class_="txt-block"):
        if i.find().text == "Gross USA:":
            return float(i.text.strip().split("\n")[0].split("$")[1].replace(",", ""))

def find_rating(soup):
    for i in soup.find(id="title-overview-widget").find_all("div", class_="ratingValue"):
        return i.text.strip().split("/")[0]
        
def find_runtime(soup):
    for i in soup.find(id="titleDetails").find_all("div", class_="txt-block"):
        if i.find().text == "Runtime:":
            return float(i.text.strip().split("\n")[1].split("min")[0].strip())
def find_date(soup):
    for i in soup.find(id="titleDetails").find_all("div", class_="txt-block"):
        if i.find().text == "Release Date:":
            return i.text.strip().split(":")[1].split("(")[0].strip()
        
def find_genre(soup):
    for i in soup.find(id="title-overview-widget").find_all("div", class_="subtext"):
        if str(i.text.strip().split("|")[2].replace('\n','').replace(' ','').split(","))[-3]==(')'):
            return i.text.strip().split("|")[1].replace('\n','').replace(' ','').split(",")
        else:
            return i.text.strip().split("|")[2].replace('\n','').replace(' ','').split(",")
        
def preparedatetime(dataframe):
    dataframe['datetime'] = pd.to_datetime(dataframe['date'], format = '%d %B %Y')
    dataframe['month'] = dataframe.datetime.dt.month
    dataframe['month'] = dataframe.datetime.dt.day
    dataframe['month'] = dataframe.datetime.dt.year
```

```python
#loop leaves us with list of alle 15 ratings in order as 1st column row 1-5, 2nd column row 1-5, 3rd column row 1-5
rate = []
for i in soup__[0].find_all('div', class_='bigcell'):
    rate.append(i.text)
rate

#explains young and old
# list_young = np.array([float(i) for i in rate]) 
# young_position = [1, 2, 6, 7, 11, 12]
# rateyoung = list(list_young[young_position])

rateyoung = list(np.array([float(i) for i in rate])[[1, 2, 6, 7, 11, 12]])
rate_young = np.array(rateyoung).sum()/6
rate_young 
rateold = list(np.array([float(i) for i in rate])[[3,4,8,9,13,14]])
rate_old = np.array(rateold).sum()/6
rate_old

rate_mean = [float(i) for i in rate][0]
rate_male = [float(i) for i in rate][5]
rate_female = [float(i) for i in rate][10]
#these should either be added to the main table, or be imported as subtable and then do inner join on title
print('mean rating:', rate_mean, 
      '\nyoung rated:', rate_young, 
      '\nold rated:', rate_old, 
      '\nmale rating:', rate_male, 
      '\nfemale rating:', rate_female)
```

```python
res = []
for row in soup__[0].find_all("tr")[1:]:
    rows = []
    for col in row.find_all("td"):
        rows.append(col.find("div").text)
    res.append(rows)
pd.DataFrame(res, columns=titles)


titless=['gender', 'all_ages', 'u18', '18_29', '30_44', 'o45']
df = pd.DataFrame(res, columns=titless)
df1 = df.set_index('gender')
df1

df1['all_ages'] = df1['all_ages'].astype(float)
df1['u18'] = df1['u18'].astype(float)
df1['18_29'] = df1['18_29'].astype(float)
df1['30_44'] = df1['30_44'].astype(float)
df1['o45'] = df1['o45'].astype(float)

df1['young'] = df1.loc[:,['u18','18_29']].sum(axis=1)/2 # - df1.loc['All','all_ages']
df1['old'] = df1.loc[:,['30_44','o45']].sum(axis=1)/2 # - df1.loc['All','all_ages']
df1
```
